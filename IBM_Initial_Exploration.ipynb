```python
# Group data by Department, Business Travel, and Attrition
df_grouped = df.groupby(['Department', 'BusinessTravel', 'Attrition']).size().unstack(fill_value=0)

# Convert counts to percentages
df_percentage = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

# Get unique departments and travel categories
departments = df_percentage.index.get_level_values(0).unique()
travel_categories = df_percentage.index.get_level_values(1).unique()

# Create subplots
fig, axes = plt.subplots(len(departments), len(travel_categories), figsize=(12, 8))

for i, dept in enumerate(departments):
    for j, travel in enumerate(travel_categories):
        ax = axes[i, j]
        
        # Get percentage data
        attrition_data = df_percentage.loc[(dept, travel)]
        
        # Plot pie chart
        ax.pie(attrition_data, labels=attrition_data.index, autopct='%1.1f%%', colors=['skyblue', 'orange'])
        ax.set_title(f"{dept}\n{travel}")

# Adjust layout
plt.suptitle("Attrition Distribution by Department & Business Travel", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Group by Department, BusinessTravel, and Attrition, then count occurrences
summary_table = df.groupby(['Department', 'BusinessTravel', 'Attrition']).size().unstack(fill_value=0)

# Convert to percentage within each Department & Business Travel group
summary_table_percentage = summary_table.div(summary_table.sum(axis=1), axis=0) * 100

# Print summary tables
print("ðŸ”¢ Count Table:")
print(summary_table)

print("\nðŸ“Š Percentage Table:")
print(summary_table_percentage)
```
